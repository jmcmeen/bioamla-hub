# iNaturalist and La Selva Research Station

Code examples using pyinaturalist to pull data for frogs at La Selva Research Station, Costa Rica.

## Getting Started

Define the place_id and taxon_id of interest from iNaturalist web application. Here we are searching for frog observations at La Selva Research Station, Coasta Rica.

```python
la_selva_place_id = 135489
frog_taxon_id = 20979
```

Print species counts given taxon id and place id from iNaturalist web app

```python
from pyinaturalist import get_observation_species_counts, TaxonCount, pprint

# get the observation species counts for frogs at Steele Creek Park
response = get_observation_species_counts(place_id= la_selva_place_id, taxon_id=frog_taxon_id)

# response comes in as a list of dictionaries in JSON format, Taxon count will help us parse it
taxa = TaxonCount.from_json_list(response['results'][:])

# pretty print out the results
pprint(taxa)
```
